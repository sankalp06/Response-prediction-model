```python
import pandas as pd

df = pd.read_csv('banking.csv')
```

```python
df.columns
```

```python
df['job'] = df['job'].map({'blue-collar':0, 'technician':1, 'management':2, 'services':3,'retired':4, 'admin.':5, 'housemaid':6, 'unemployed':7, 'entrepreneur':8, 'self-employed':9, 'unknown':10, 'student':11}).astype(int) 
df['marital'] = df['marital'].map({'married':0, 'single':1, 'divorced':2, 'unknown':3}).astype(int)
df['education'] = df['education'].map({'basic.4y':0, 'unknown':1, 'university.degree':2, 'high.school':3, 'basic.9y':4,'professional.course':5, 'basic.6y':6, 'illiterate':7}).astype(int)
df['default'] = df['default'].map({'unknown':0, 'no':1, 'yes':2}).astype(int)
df['housing'] = df['housing'].map({'unknown':0, 'no':1, 'yes':2}).astype(int)
df['loan'] = df['loan'].map({'unknown':0, 'no':1, 'yes':2}).astype(int)
df['contact'] = df['contact'].map({'cellular':0, 'telephone':1}).astype(int)
df['month'] = df['month'].map({'aug':0, 'nov':1, 'jun':2, 'apr':3, 'jul':4, 'may':5, 'oct':6, 'mar':7, 'sep':8,
       'dec':9,'feb':10,'apr':11}).astype(int)
df['day_of_week'] = df['day_of_week'].map({'thu':3, 'fri':4,'sat':5, 'sun':6, 'tue':1, 'mon':0, 'wed':2}).astype(int)
df['poutcome'] = df['poutcome'].map({'nonexistent':0, 'success':1, 'failure':2}).astype(int)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import precision_score
from sklearn.metrics import confusion_matrix

y = df['y']
X = df.drop('y', axis =1)


x_train_orginal, x_test_orginal, y_train, y_test = train_test_split(X, y, random_state = 42, test_size = 0.2)
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train_orginal)
x_test = scaler.fit_transform(x_test_orginal)
```

```python
x_test
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression


model = LogisticRegression(max_iter=300,n_jobs=-1) 

#TRAINING THE MODEL
model.fit(x_train,y_train)

#TESTING THE MODEL
confidence = model.score(x_test,y_test)


print(confidence)

```

# SVC - LINEAR

```python
from sklearn.svm import SVC

svc_linear = SVC(kernel = 'linear', C = 10, gamma = 0.1)
svc_linear.fit(x_train, y_train)

confidence = svc_linear.score(x_test,y_test)


print(confidence)
```

# SVC - Rbf

```python
from sklearn.svm import SVC

svc_linear = SVC(kernel = 'rbf', C = 100, gamma = 0.1)
svc_linear.fit(x_train, y_train)

confidence = svc_linear.score(x_test,y_test)


print(confidence)
```
